```python
def entropy (col):
    ent = 0.0
    counts = np.unique(col, return_counts = True)
    for x in counts[1]:
        pi = x/ len(col)
        ent += -1*pi*np.log2(pi)
    return ent
```

```python
def info_gain (df, attr, target):
    tot = entropy(df(target)) 
    val = np.unique(df[attr])
    acc = 0
    for x in val:
        subset = df[df[attr] == x][target]
        x = len(subset) / len(df)
        acc += x * entropy(subset)
    return tot - acc
    
    
```

```python
Explicação do Código Info:
    -Tot equivale ao valor da entropia de nosso "Target" que nesse caso é "Survived"
    -val equivale aos valores únicos de um atributo
    - No for buscamos x em val, uma possivel exemplificação seria:
        x = [male], val = [male,female]
    -subset equivale a uma lista de todas as linhas onde nosso "Target" se relaciona com val
    
```
